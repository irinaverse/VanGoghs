"Head of a Skeleton with a Burning Cigarette" is a Vincent Van Gogh painting, dated to January-February 1886. He painted it while studying at the Art Academy in Antwerp, Belgium. Drawing skeletons was a standard exercise at the academy, but painting them was not part of the curriculum. He must have made it at some other time, between or after his lessons.

```python
skeleton_image = imread("skull with burning cigarette.jpg")
plt.figure(figsize = (4, 7))
plt.title("Head of a Skeleton with a Burning Cigarette", size = 9)
plt.axis("off")
plt.imshow(skeleton_image)
plt.show()
```

In modern $Image$ $Processing$ Vincent's painting is an image, arranged in columns and rows, as a matrix. The matrix contains pixels (picture elements).

```python
skeleton_image.shape        # width, height, color channels
```

```python
# manually
start_point = (1400, 1450)            # (x-axis, y-axis)
end_point = (1700, 1670)            # (x-axis, y-axis)
line_color = (0, 255, 0)          # RGB model
line_thickness = 3

cigarette_on_image = cv2.rectangle(skeleton_image, start_point, end_point, line_color, line_thickness)
plt.xlabel("$x$", loc = "right")
plt.ylabel("$y$", loc = "top", rotation = "horizontal")
plt.imshow(cigarette_on_image)

plt.show()
```

The spatial coordinates $x$ and $y$ enable us to specify a location in the image. They are horizontal and vertical addresses in a 2D space. The $x$-axis always comes first, followed by the $y$-axis. Above you see how we detect an object manually, pointing parts on an image. It is useful in medical images, places, maps and whenever needed.
